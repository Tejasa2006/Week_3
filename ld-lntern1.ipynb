```python
a = df['AMT_INCOME_TOTAL']
plt.boxplot(a);
Q3 = np.percentile(a,75)
Q1 = np.percentile(a,25)
iqr = Q3-Q1
iqr,Q3,Q1
```

```python
ul = Q3 + (1.5 * iqr)
lo = Q1 - (1.5 * iqr)
```

```python
ul,lo
```

```python
df = df[(df['AMT_INCOME_TOTAL'] <= ul) & (df['AMT_INCOME_TOTAL'] >= lo)]
```

```python
sns.boxplot(df['AMT_INCOME_TOTAL']);
plt.title('Outliers in income')
```
