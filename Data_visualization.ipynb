```python
# To merge the 2 plots 

plt.title('Line plot')
plt.xlabel('First Value')
plt.ylabel('Second value')# Marker gives it a point
plt.xticks(range(1,90,5))  # To give the range of the x axis
plt.yticks(range(1,90,5))   # To give the range of the y axis
plt.plot(x,y,marker = 'o',mfc = 'red',color = 'black',ls = '--' ,
         ms = 5,label = '2023'); 
plt.plot(x,z,marker = 'o',mfc = 'yellow',color = 'purple',ls = '--' ,
         ms = 5,label = '2024'); 
plt.grid()
plt.legend(); # Addes the sacle to the graphs
```

## Pie Chart

## Box plot

```python
marks = [67,65,55,59,69,72,51,70,53,98,23]
```

```python
plt.boxplot(marks);
plt.grid()
```

```python
Q3 = np.percentile(marks,75)
Q1 = np.percentile(marks,25)
Q3,Q1
```

```python
iqr = Q3-Q1
iqr         # To find the inter qrartile range
```

```python
ub = Q3 + (1.5 * iqr)  # To calculate the upper outlier
lb = Q1 - (1.5 * iqr)  # To calculate the lower outlier
```
